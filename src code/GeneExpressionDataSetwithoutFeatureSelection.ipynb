```python
from sklearn.impute import SimpleImputer 
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
#from sklearn.preprocessing import Imputer
#imputer=Imputer(missing_values = np.nan,strategy='mean', axis=0,verbose=0)
imputer.fit(x)
x = imputer.transform(x)
x
```

```python
column_names=list(data.columns)[1:len(list(data.columns))-1]
```

```python
test_df=pd.DataFrame(x,columns=column_names)
```

```python
test_df
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(test_df, y, test_size=0.25, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB 
gnb = GaussianNB() 
gnb.fit(X_train, y_train) 
y_pred = gnb.predict(X_test) 
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score 
print ("Accuracy NB: ", accuracy_score(y_test, y_pred)) 
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=27, metric='euclidean')
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score 
print ("Accuracy knn: ", accuracy_score(y_test, y_pred)) 
```

```python
# Create Decision Tree classifer object
from sklearn.tree import DecisionTreeClassifier 
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score 
print ("Accuracy decision: ", accuracy_score(y_test, y_pred)) 
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=5)
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score 
print ("Accuracy Random forest: ", accuracy_score(y_test, y_pred)) 
```

```python
from sklearn import svm
from sklearn.multiclass import OneVsRestClassifier
clf = OneVsRestClassifier(svm.SVC(gamma=0.001, C=10., probability=True, class_weight='balanced', kernel='linear'))
clf.fit(X_train, y_train)
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score 
print ("Accuracy : ", accuracy_score(y_test, y_pred)) 
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
```

```python
learning_rates = [0.05, 0.1, 0.25, 0.5, 0.75, 1]
for learning_rate in learning_rates:
    gb = GradientBoostingClassifier(n_estimators=20, learning_rate = learning_rate, max_features=2, max_depth = 2, random_state = 0)
    gb.fit(X_train, y_train)
    print("Learning rate: ", learning_rate)
    print("Accuracy score (training): {0:.3f}".format(gb.score(X_train, y_train)))
    print("Accuracy score (validation): {0:.3f}".format(gb.score(X_test, y_test)))
    print()
```

```python
gb = GradientBoostingClassifier(n_estimators=20, learning_rate = 0.5, max_features=2, max_depth = 2, random_state = 0)
gb.fit(X_train, y_train)
predictions = gb.predict(X_test)

print("Confusion Matrix:")
print(confusion_matrix(y_test, predictions))
print()
print("Classification Report")
print(classification_report(y_test, predictions))
from sklearn.metrics import accuracy_score 
print ("Accuracy : ", accuracy_score(y_test, predictions)) 
```
